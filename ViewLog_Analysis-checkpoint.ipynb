```python
log1 = 'anonym_viewthread-20181130.log'
log2 = 'anonym_viewthread-20181203.log'
log3 = 'anonym_viewthread-20181206.log'

lst_20181130 = []
lst_20181203 = []
lst_20181206 = []
```

```python
# Read in json to list
def read_data(log):
    output_lst=[]
    for line in open(log).readlines():
        output_lst.append(json.loads(line))
    output_lst
    return output_lst
#    print(len(feeds_read_from_file))
```
